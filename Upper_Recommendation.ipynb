```python
def processing(df_dummy):
    for column in df_dummy.columns:
        if '/' in column and 'Color' in column:
            color = column.split('/')
            for i_row in range(len(df_dummy[column])):
                if df_dummy[column][i_row] == 1:
                    if color[0] in df_dummy.columns:
                        df_dummy[color[0]][i_row]=1
                    else:
                        df_dummy[color[0]] = 0
                        df_dummy[color[0]][i_row]=1

                    if ('Color_'+color[1]) in df_dummy.columns:
                        df_dummy['Color_'+color[1]][i_row]=1
                    else:
                        df_dummy['Color_'+color[1]] = 0
                        df_dummy['Color_'+color[1]][i_row]=1
        
        if '/' in column and 'Style' in column:
            style = column.split('/')
            for i_row in range(len(df_dummy[column])):
                if df_dummy[column][i_row] == 1:
                    if style[0] in df_dummy.columns:
                        df_dummy[style[0]][i_row]=1
                    else:
                        df_dummy[style[0]] = 0
                        df_dummy[style[0]][i_row]=1

                    if ('Style_'+style[1]) in df_dummy.columns:
                        df_dummy['Style_'+style[1]][i_row]=1
                    else:
                        df_dummy['Style_'+style[1]] = 0
                        df_dummy['Style_'+style[1]][i_row]=1

    for column in df_dummy.columns:
        if '/' in column:
            df_dummy.drop(columns=[column], inplace = True)

    top_attr_columns = top_attr_dummy.columns[1:]
    df_dummy = df_dummy.reindex(columns=['FileName'] + sorted(top_attr_columns))

    return df_dummy
                
    
```

# Evaluate

```python
def check_similar(item1, item2, len_column):
    false_count = 0
    if item1[-1] not in item2[-1]:
        return 'EventFalse'
    else:
        for i in range(1,len_column-1):
            
            if '/' not in item1[i] :
                if item1[i] not in item2[i]:
                    false_count += 1
            else:
                options = item1[i].split('/')
                if options[0] not in item2[i] and options[1] not in item2[i]:
                    false_count += 1
        
    if false_count/(len_column-2) == 0:
        return "Completely"
    else:
        if false_count/(len_column-2) <= 0.2:
            return "Relatively"
        else:
            return "Under80%"
    
```

```python
def score_a_result(item,results_df, results):
    event_true = len(results_df)
    completely = 0
    relatively = 0 #similar 80%-100%
    for index,row in results_df.iterrows():
        if check_similar(item, row, len(results_df.columns)) == 'Completely':
            completely+=1
        if check_similar(item, row, len(results_df.columns)) == 'Relatively':
            relatively+=1
        if check_similar(item, row, len(results_df.columns)) == 'EventFalse':
            event_true-=1
    return completely/len(results_df), (relatively+completely)/len(results_df), event_true/len(results_df)
                    
```

```python
def evaluate(k):
    completely_score = []
    relatively_score = []
    event_true_score = []
    for index, item in top_attr_dummy.iloc[:, 1:].iterrows():
        cos_sim = cosine_similarity(top_attr_dummy.iloc[:, 1:], np.array(item).reshape(1,-1))
        cos_sim = pd.DataFrame(data=cos_sim, index=top_attr_dummy.index)
        cos_sim.sort_values(by = 0, ascending = False, inplace=True)
        results = cos_sim.index.values[1:k+1]
        results_df = top_attr.loc[results]
        score = []
        score = score_a_result(top_attr.iloc[index],results_df, results)
        completely_score.append(score[0])
        relatively_score.append(score[1])
        event_true_score.append(score[2])
    return np.mean(completely_score), np.mean(relatively_score), np.mean(event_true_score)
```

```python
k = 1
score = evaluate(4)
print("With {} suggestion(s) , similar on all attributes: {}".format(k, score[0]))
print("With {} suggestion(s) , similar over 80 percents of attributes: {}".format(k, score[1]))
print("With {} suggestion(s) , similar on event label: {}".format(k, score[2]))
```
